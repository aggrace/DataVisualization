```python
trace0=go.Scatter(
    x=beijing2["followers"],
    y=beijing2["district"],
    name="housing",
    mode="markers",
    marker=dict(
    size=10,
        color='rgba(152,0,0,0.8)',
        line=dict(
        width=2,
            color='rgb(0,0,0)'
        )
    )

)

```

```python
trace1=go.Scatter(
    x=beijing2["followers"],
    y=beijing2["ladderRatio"],
    name="ladder",
    mode="markers",
    marker=dict(
    size=10,
        color='rgba(255,182,193,0.9)',
        line=dict(
        width=2
        )
    )

)
```

```python
data=[trace0,trace1]
layout=dict(title='Housing scatters',
           
           xaxis=dict(
               title="followers",
            titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
            )
           ),
            
            yaxis=dict(
            title='district/Ladder ratio',
            titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
            )
            )
           )

fig=go.Figure(data=data,layout=layout)
#plotly.offline.plot(fig,filename='housing scatter.html')
```
